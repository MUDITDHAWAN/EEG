```python
import os
import numpy as np
import pandas as pd
```

```python
os.chdir('/home/shreya/DEAP/CNN/TaskValence/HighValence')
high= []
high = [[] for ind in range(33)]
for file in os.listdir('/home/shreya/DEAP/CNN/TaskValence/HighValence'):
    if(file[0] == 's'):
#         print(file)
        subject = file[1:3]
#         print(subject)
        if subject[0] == '0':
            subj_no = int(subject[1])
        else:
            subj_no = int(subject)
#         print(type(subj_no))
        trial = file[4:6]
#         print(trial)
        if trial[1] == '_':
            trial_no = int(trial[0])
        else:
            trial_no = int(trial)
#         print(type(trial_no))
        high[subj_no].append(trial_no)
#         break
```

```python
os.chdir('/home/shreya/DEAP/CNN/TaskValence/LowValence')
low= []
low = [[] for ind in range(33)]
for file in os.listdir('/home/shreya/DEAP/CNN/TaskValence/LowValence'):
    if(file[0] == 's'):
#         print(file)
        subject = file[1:3]
#         print(subject)
        if subject[0] == '0':
            subj_no = int(subject[1])
        else:
            subj_no = int(subject)
#         print(type(subj_no))
        trial = file[4:6]
#         print(trial)
        if trial[1] == '_':
            trial_no = int(trial[0])
        else:
            trial_no = int(trial)
#         print(type(trial_no))
        low[subj_no].append(trial_no)
#         break
```

```python
os.chdir('/home/shreya/DEAP/CNN/TaskValence/NoneValence')
none= []
none = [[] for ind in range(33)]
for file in os.listdir('/home/shreya/DEAP/CNN/TaskValence/NoneValence'):
    if(file[0] == 's'):
#         print(file)
        subject = file[1:3]
#         print(subject)
        if subject[0] == '0':
            subj_no = int(subject[1])
        else:
            subj_no = int(subject)
#         print(type(subj_no))
        trial = file[4:6]
#         print(trial)
        if trial[1] == '_':
            trial_no = int(trial[0])
        else:
            trial_no = int(trial)
#         print(type(trial_no))
        none[subj_no].append(trial_no)
#         break
```

```python
high = [ list(set(trial_list)) for trial_list in high]
low = [ list(set(trial_list)) for trial_list in low]
none = [ list(set(trial_list)) for trial_list in none]
```

```python
df_trials = pd.DataFrame(list(zip(high, low, none)), columns =['High', 'Low', 'None'])
```

```python
df_trials
```

```python
high_nb = [ len(trial_list) for trial_list in high]
low_nb = [ len(trial_list) for trial_list in low]
none_nb= [ len(trial_list) for trial_list in none]
```

```python
df_nb = pd.DataFrame(list(zip(high_nb, low_nb, none_nb)), columns =['High', 'Low', 'None'])
```

```python
os.chdir('/home/shreya/DEAP/CNN')
df_nb.to_csv('Tv.csv')
```
